```python
import base64
import io
import matplotlib.pyplot as plt
import cv2 as cv
import numpy as np
```

```python
# request data
im = '/home/irizqy/ml_ws/bangkit-ws/data/temp/jj_6.jpeg'
# im = '/home/irizqy/ml_ws/bangkit-ws/data/bizit-dev_test-data/32.jpg'
# im = '/home/irizqy/Downloads/rfc.jpg'

with open(im, "rb") as f:
    im_bytes = f.read()        
im_b64 = base64.b64encode(im_bytes).decode("utf8")
print(type(im_b64))
```

```python
jpg_original = base64.b64decode(im_b64.encode('utf-8'))
jpg_as_np = np.frombuffer(jpg_original, dtype=np.uint8)
img = cv.imdecode(jpg_as_np, flags=1)
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
from PIL import Image

img_bytes = base64.b64decode(im_b64.encode('utf-8'))
im = Image.open(io.BytesIO(img_bytes)).resize((512, 512))
im_arr = np.asarray(im)

plt.imshow(im_arr)
```
